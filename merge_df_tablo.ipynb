```python
#merging the 2 df using web_trans_sum as base and bringing the customer info from Vanguard_ab
df_ab_tablo = web_tran_sum.merge(Vanguard_ab, on='client_id', how='left')
```

```python
#dropping na in the final df
df_ab_tablo_clean = df_ab_tablo.dropna(subset=['clnt_age'])
```

```python
# Convert to an excel file
df_ab_tablo_clean.to_excel('AB_testing_tablo_final.xlsx', index=True)
```

```python
# Create a column 'is_confirmed' that checks if 'transaction_status' contains 'confirmed'
df_ab_tablo_clean['is_confirmed'] = df_ab_tablo_clean['transaction_status'].str.contains('confirmed', case=False, na=False)

# Calculate the completion rate for Test and Control populations
completion_rate = df_ab_tablo_clean.groupby('variation')['is_confirmed'].mean() * 100

# Display the completion rate
print(completion_rate)
```

```python
# Calculate total number of transactions per variation
total_transactions = df_ab_tablo_clean.groupby('variation')['visit_id'].count()

# Calculate the number of transactions where 'is_confirmed' is True
confirmed_transactions = df_ab_tablo_clean.groupby('variation')['is_confirmed'].sum()

# Create a DataFrame to show the results
result_df = pd.DataFrame({
    'Total Transactions': total_transactions,
    'Confirmed Transactions': confirmed_transactions,
    'Completion Rate': (confirmed_transactions / total_transactions) * 100
})

# Format the completion rate to display with 2 decimal places
result_df['Completion Rate'] = result_df['Completion Rate'].round(2)

# Reset the index to remove the 'variation' label as an index
result_df_reset = result_df.reset_index()

# Display the result
result_df_reset
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Number of confirmed transactions for each group
confirmed = [result_df.loc['Test', 'Confirmed Transactions'], result_df.loc['Control', 'Confirmed Transactions']]

# Total number of transactions for each group
nobs = [result_df.loc['Test', 'Total Transactions'], result_df.loc['Control', 'Total Transactions']]

# Perform Z-test for proportions
z_stat, p_value = proportions_ztest(confirmed, nobs)

# Display results
print(f"Z-statistic: {z_stat}, p-value: {p_value}")

#H0​: pTest ​= pControl​
#H1​: pTest ​!= pControl​
#With p = Number of confirmed transactions / Number of Total transactions

#Since the p-value is so low, we reject the null hypothesis (H₀). 
#This means that the difference in completion rates between the Test and Control groups is statistically significant
#The variation introduced in the Test group has had a statistically significant impact on the completion rate.

```

```python
# Average Time Spent on Each Step
# Define the step columns
step_columns = ['start_duration', 'step1_duration_exact', 'step2_duration_exact', 'step3_duration_exact', 'confirm_duration_exact','total_duration']

# Calculate the average duration per step for each group (Test and Control)
average_durations = df_ab_tablo_clean.groupby('variation')[step_columns].mean().round(1)

# Rename the columns
average_durations = average_durations.rename(columns={
    'start_duration': 'Start',
    'step1_duration_exact': 'Step 1',
    'step2_duration_exact': 'Step 2',
    'step3_duration_exact': 'Step 3',
    'confirm_duration_exact': 'Confirmed',
    'total_duration': 'Total'
})

# Calculate the difference between Test and Control for each column
difference_row = average_durations.loc['Test'] - average_durations.loc['Control']

# Append the difference as a new row in the DataFrame
average_durations.loc['Difference'] = difference_row

# Reset the index to remove the current index (variation) and make it part of the columns
average_durations_reset = average_durations.reset_index(drop=True)

# Define a function to apply color formatting based on positive and negative values
def color_positive_negative(val):
    color = 'red' if val > 0 else 'green' if val < 0 else 'black'
    return f'color: {color}'

# Apply the color formatting to the 'Difference' row
styled_df = average_durations.style.applymap(color_positive_negative, subset=pd.IndexSlice['Difference', :])
styled_df = styled_df.format("{:.1f}")

# Display the summary in a DataFrame
styled_df
```

```python
from statsmodels.stats.weightstats import ztest

# Define the step columns
step_columns = ['start_duration', 'step1_duration_exact', 'step2_duration_exact', 'step3_duration_exact', 'confirm_duration_exact', 'total_duration']

# Dictionary to store the Z-test results
z_test_results = {}

# Loop over each step and perform the Z-test for Test and Control groups
for step in step_columns:
    # Extract the Test and Control groups for the current step
    test_group = df_ab_tablo_clean[df_ab_tablo_clean['variation'] == 'Test'][step]
    control_group = df_ab_tablo_clean[df_ab_tablo_clean['variation'] == 'Control'][step]
    
    # Perform the Z-test
    z_stat, p_value = ztest(test_group, control_group)
    
    # Store the Z-statistic and p-value for the current step
    z_test_results[step] = {'Z-statistic': z_stat, 'p-value': p_value}

# Convert the Z-test results to a DataFrame for display
z_test_df = pd.DataFrame(z_test_results).T

# Rename the rows (index) of the Z-test results DataFrame for clearer presentation
z_test_df_renamed = z_test_df.rename(index={
    'start_duration': 'Start',
    'step1_duration_exact': 'Step 1',
    'step2_duration_exact': 'Step 2',
    'step3_duration_exact': 'Step 3',
    'confirm_duration_exact': 'Confirmed',
    'total_duration': 'Total'
})


# Display the Z-test results
z_test_df_renamed
```

```python
#H0​:mu Test​ = mu Control​
#H1​:mu Test != mu Control​
```

```python
# Define the comments for each row based on the Z-test results
comments = [
    "YES",
    "NO",
    "YES",
    "NO",
    "YES",
    "YES",
]

# Add the comments as a new column to the DataFrame
z_test_df_renamed['Statiscally significant difference'] = comments

# Display the updated DataFrame with conclusions
z_test_df_renamed
```

```python
from statsmodels.stats.weightstats import ztest

# Filter the dataset to include only rows where 'transaction_status' is confirmed (excluding errors)
df_confirmed = df_ab_tablo_clean[df_ab_tablo_clean['transaction_status'].str.contains('confirmed', case=False, na=False)]

# Extract the repetition values for Test and Control groups where the transaction status is confirmed
test_group_repetition_confirmed = df_confirmed[df_confirmed['variation'] == 'Test']['Repetition']
control_group_repetition_confirmed = df_confirmed[df_confirmed['variation'] == 'Control']['Repetition']

# Perform the Z-test for Repetition between Test and Control groups
z_stat_repetition_confirmed, p_value_repetition_confirmed = ztest(test_group_repetition_confirmed, control_group_repetition_confirmed)

# Display the Z-test results
print(f"Z-statistic: {z_stat_repetition_confirmed}, p-value: {p_value_repetition_confirmed}")
```
